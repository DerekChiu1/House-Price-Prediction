#### Purpose of the file:
The purpose of this file is to process and clean the dataset, making the data ready for use in machine learning, visualization, and analysis. The cleaned dataset will be saved to another csv file, which will be used for future work and be uploaded to the project repository on GitHub as well.

```python
# Import all needed libraries

import numpy as np
import pandas as pd
```

```python
# Load the dataset into a data frame

house_price_df = pd.read_csv("HousePricePrediction.csv")
print (f"Shape of the data frame: {house_price_df.shape}\n")
house_price_df.head()
```

```python
# Remove all rows with one or more NaN entries

house_price_df.dropna(inplace=True)
print (f"Shape of the data frame after dropping rows containing NaN entries: {house_price_df.shape}\n")
```

```python
# Reset the Id column

house_price_df["Id"] = range(1, house_price_df.shape[0]+1)
house_price_df
```

```python
# Remove irrelevant columns

irrelevant_columns = ["MSZoning", "Exterior1st", "BsmtFinSF2", "YearRemodAdd"]
house_price_df.drop(irrelevant_columns, axis=1, inplace=True)
house_price_df.head()
```

```python
# Use one-hot encoding to trnsform categorical features into numerical features

categorical_columns = ["LotConfig", "BldgType"]
house_price_df = pd.get_dummies(house_price_df, columns=categorical_columns, drop_first=True, dtype=float)
print (f"Shape of the data frame after one-hot encoding: {house_price_df.shape}\n")
house_price_df.head()
```

```python
# Save the cleaned data frame into a new csv file

house_price_df.to_csv("Cleaned_house_price.csv", index=False)
```
